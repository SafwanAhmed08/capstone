```python
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset/processed_dataset.csv")  

seen_attacks = set()

for idx, row in df.iterrows():
    attack_type = row.get("Attack_type")

    if attack_type in seen_attacks:
        continue

    seen_attacks.add(attack_type)

    print(f"\n=== First Row of Attack: {attack_type} (Row {idx}) ===")

    for col, val in row.items():
        if pd.notnull(val) and str(val).strip() not in ['0', '0.0'] and val != 0:
            print(f"{col}: {val}")

    print("=" * 50)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Selected dataset for ML and DL/DNN-EdgeIIoT-dataset.csv")  # Adjust path as needed

# Dictionary to store non-null columns for each attack type
attack_feature_map = {}

# Group by attack type
for attack_type, group in df.groupby("Attack_type"):
    # Get only columns that have at least one non-null, non-zero value
    meaningful_cols = [
        col for col in group.columns
        if col not in ['Attack_type', 'Attack_label'] and  # optional: exclude label columns
           group[col].apply(lambda x: pd.notnull(x) and str(x).strip() not in ['0', '0.0'] and x != 0).any()
    ]
    
    attack_feature_map[attack_type] = meaningful_cols

# Print results
for attack, features in attack_feature_map.items():
    print(f"\n=== Attack Type: {attack} ===")
    for f in features:
        print(f"- {f}")
    print(f"Total meaningful features: {len(features)}")
    print("=" * 60)
```
